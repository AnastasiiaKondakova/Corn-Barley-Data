```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = [ 'Date',
            'Corn Price',
            'Barley Price',
           	'Wheat Price',
            'CPI',
           	'GPR	Country',
            'GPR',
           	'US Wheat Futures',
           	'Brent Oil Futures',
           	'London Wheat Futures',
           	'Exch Rate']
df = pd.read_excel('EDA.xlsx', header = None, skiprows = 1, names = columns)
```

```python
plt.figure(figsize=(10,6))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
```

```python
import pandas as pd
 
xls = pd.ExcelFile('Climate data.xlsx')
sheet_names = xls.sheet_names
del sheet_names[0]
```
