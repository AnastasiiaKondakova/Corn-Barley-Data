```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np
import datetime
```

```python
columns_corn = [ 'Date',
            'Corn Price',
            'CPI',
           	'GPR	Country',
            'GPR',
           	'US Wheat Futures',
           	'Brent Oil Futures',
           	'London Wheat Futures',
           	'Exch Rate',
             'T',
             'TM',
             'Tm',
             'H',
             'PP',
           	 'VV',
             'V',
             'VM',
             'RA',
           	 'SN',
             'TS',
             'FG'  ]

columns_barley = [ 'Date',
            'Barley Price',
            'CPI',
           	'GPR	Country',
            'GPR',
           	'US Wheat Futures',
           	'Brent Oil Futures',
           	'London Wheat Futures',
           	'Exch Rate',
             'T',
             'TM',
             'Tm',
             'H',
             'PP',
           	 'VV',
             'V',
             'VM',
             'RA',
           	 'SN',
             'TS',
             'FG'  ]

df_corn = pd.read_excel('Corn Data.xlsx', header = None, skiprows = 1, names = columns_corn)
df_barley = pd.read_excel('Barley Data.xlsx', header = None, skiprows = 1, names = columns_barley)
```

```python
df_corn.index = df_corn["Date"]
df_corn.index = pd.to_datetime(df_corn.index)
df_corn.drop(['Date'], axis='columns', inplace=True)
df_corn = df_corn.sort_index()
```

```python
df_barley.index = df_barley["Date"]
df_barley.index = pd.to_datetime(df_barley.index)
df_barley.drop(['Date'], axis='columns', inplace=True)
df_barley = df_barley.sort_index()
```

```python
df_corn["Exchange Rate Change 1"] = df_corn["Exch Rate"].pct_change(periods=1)*100
df_corn["Exchange Rate Change 4"] = df_corn["Exch Rate"].pct_change(periods=4)*100
```

```python
df_barley["Exchange Rate Change 1"] = df_barley["Exch Rate"].pct_change(periods=1)*100
df_barley["Exchange Rate Change 4"] = df_barley["Exch Rate"].pct_change(periods=4)*100
```

```python
df_corn = df_corn.dropna()
df_barley = df_barley.dropna()
```

```python
X_train_corn, X_test_corn = train_test_split(df_corn, train_size=0.8, shuffle=False)
X_train_barley, X_test_barley = train_test_split(df_barley, train_size=0.8, shuffle=False)
```

#**Corn**

#CART

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as MSE

SEED = 1
```

```python
max_depth_=4
min_samples_leaf_=0.08

dt = DecisionTreeRegressor(max_depth=max_depth_, min_samples_leaf=min_samples_leaf_, random_state=1)
model_dt=dt.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_train_corn)
rmse=np.sqrt(mean_squared_error(y_train_corn,y_dt_pred))
print('max_depth='+str(max_depth_)+' min_samples_leaf_='+str(min_samples_leaf_)+' RMSE='+str(rmse))
```

```python
max_depth_=4
min_samples_leaf_=0.08

dt = DecisionTreeRegressor(max_depth=max_depth_, min_samples_leaf=min_samples_leaf_, random_state=1)
model_dt=dt.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_test_corn)
rmse=np.sqrt(mean_squared_error(y_test_corn,y_dt_pred))
print('max_depth='+str(max_depth_)+' min_samples_leaf_='+str(min_samples_leaf_)+' RMSE='+str(rmse))
```

```python
corn_results_test = pd.DataFrame(y_test_corn.copy())
corn_results_test['pred CART'] = pd.DataFrame(y_dt_pred)[0].values
corn_results_test
```

#RF

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.ensemble import RandomForestRegressor
```

```python
n_estimators_=150 
max_features_='auto' 
min_samples_leaf_=5

dt = RandomForestRegressor(n_estimators = n_estimators_, max_features = max_features_, min_samples_leaf = min_samples_leaf_, n_jobs=-1, random_state=2) 
model_dt=dt.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_train_corn)
rmse=np.sqrt(mean_squared_error(y_train_corn,y_dt_pred))
print('n_estimators='+str(n_estimators_)+' max_features='+str(max_features_)+' min_samples_leaf='+str(min_samples_leaf_)+' RMSE='+str(rmse))
```

```python
n_estimators_=150 
max_features_='auto' 
min_samples_leaf_=5

dt = RandomForestRegressor(n_estimators = n_estimators_, max_features = max_features_, min_samples_leaf = min_samples_leaf_, n_jobs=-1, random_state=2) 
model_dt=dt.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_test_corn)
rmse=np.sqrt(mean_squared_error(y_test_corn,y_dt_pred))
print(': n_estimators='+str(n_estimators_)+' max_features='+str(max_features_)+' min_samples_leaf='+str(min_samples_leaf_)+' RMSE='+str(rmse))
```

```python
corn_results_test['pred RF'] = pd.DataFrame(y_dt_pred)[0].values
corn_results_test
```

#kNN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
scaler_corn = StandardScaler()
scaler_corn.fit(X_train_corn)
X_train_corn = scaler_corn.transform(X_train_corn)
X_test_corn = scaler_corn.transform(X_test_corn)
```

```python
knn=KNeighborsRegressor(n_neighbors=2, algorithm = 'brute')
model_knn=knn.fit(X_train_corn,y_train_corn)
y_knn_pred=model_knn.predict(X_train_corn)
rmse=np.sqrt(mean_squared_error(y_train_corn,y_knn_pred))
rmse
```

```python
knn=KNeighborsRegressor(n_neighbors=2, algorithm = 'brute')
model_knn=knn.fit(X_train_corn,y_train_corn)
y_knn_pred=model_knn.predict(X_test_corn)
rmse=np.sqrt(mean_squared_error(y_test_corn,y_knn_pred))
rmse
```

```python
corn_results_test['pred kNN'] = pd.DataFrame(y_knn_pred)[0].values
corn_results_test
```

#SVR

```python
from sklearn.svm import SVR
```

```python
svr_1 = SVR(kernel="linear", C=500)
model_dt=svr_1.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_train_corn)
rmse=np.sqrt(mean_squared_error(y_train_corn,y_dt_pred))
print(' RMSE='+str(rmse))
```

```python
svr_1 = SVR(kernel="linear", C=500)
model_dt=svr_1.fit(X_train_corn,y_train_corn)
y_dt_pred = model_dt.predict(X_test_corn)
rmse=np.sqrt(mean_squared_error(y_test_corn,y_dt_pred))
print(' RMSE='+str(rmse))
```

```python
corn_results_test['pred SVR'] = pd.DataFrame(y_dt_pred)[0].values
corn_results_test
```

```python
fig,ax=plt.subplots(figsize=(10,10))

ax.plot(corn_results_test.index,corn_results_test['pred SVR'], color='green', linewidth=1, label = 'SVR')
ax.plot(corn_results_test.index,corn_results_test['Corn Price'], color='orange', linewidth=2, label = 'Фактична ціна')

ax.set_xlabel('Дата', fontsize=20)
ax.set_ylabel('Ціна на кукурудзу' ,fontsize=20)

ax.legend(loc='upper left', fontsize=15)
```

```python
fig,ax=plt.subplots(figsize=(20,10))

ax.plot(corn_results_test.index,corn_results_test['pred RF'], color='red', linewidth=1, label = 'RF')
ax.plot(corn_results_test.index,corn_results_test['pred CART'], color='blue', linewidth=1, label = 'CART')
ax.plot(corn_results_test.index,corn_results_test['pred kNN'], color='black', linewidth=1, label = 'kNN')
ax.plot(corn_results_test.index,corn_results_test['pred SVR'], color='green', linewidth=1, label = 'SVR')
ax.plot(corn_results_test.index,corn_results_test['Corn Price'], color='orange', linewidth=2, label = 'Фактична ціна')

ax.set_xlabel('Дата', fontsize=20)
ax.set_ylabel('Ціна на кукурудзу' ,fontsize=20)

ax.legend(loc='upper left', fontsize=15)
```
